```python
# we use (to_datetime) to initialize the date because system not know that this is given date by user.
df["Date of Occurrence"]=pd.to_datetime(df["Date of Occurrence"])
```

```python
df.info()
```

```python
# We create the column of month and month is given according to date of occurrence columns.
df["Months"]=df["Date of Occurrence"].dt.month_name()
```
